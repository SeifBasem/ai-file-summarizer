```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
texts = [
    "I love this movie!",
    "This is the worst book I've ever read.",
    "The food was absolutely fantastic and the service was excellent.",
    "I’m really disappointed with the product quality.",
    "What a wonderful experience, I’ll definitely come back again.",
    "The app keeps crashing and it’s so frustrating.",
    "I feel neutral about this — it’s okay, but nothing special.",
    "The concert last night was unforgettable!",
    "Terrible customer support, I won’t recommend this company.",
    "The new update made the software much better."
]
```

```python

results = classifier(texts)


for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}, Score: {result['score']:.4f}")
    print()
```
